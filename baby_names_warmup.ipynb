# Baby Names

This dataset contains information on how many babies were born with each name. Both the raw number and proportion of births for the year are included.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

url = 'https://raw.githubusercontent.com/zgulde/tidytuesday/master/data/2022/2022-03-22/babynames.csv'
df = pd.read_csv(url)
df.head()
```

- What were the 10 most popular names the year you were born?
- What year was your name the most popular?
- Where does your name rank for the year that you were born?
- Does your name increase or decrease in popularity compared to the year after the year you were born? The year before? What about 5 years before/after?
- Visualize your name's popularity over time.

```python
# What were the 10 most popular names the year you were born?
df_1992 = df[df['year'] == 1992].sort_values('n', ascending=False).reset_index(drop=True)
df_1992.head(10)
```

```python
# What year was your name the most popular? Craig.
# reset index
df_craig = df[df['name'] == 'Craig'].sort_values('n', ascending=False).reset_index()
df_craig.head()
```

```python
# Where the name 'Craig' rank for the year 1992?
df_1992[df_1992['name'] == 'Craig']
```

```python
# Does 'Craig' increase or decrease in popularity compared to the year after the year you were born?
name_1993 = df[df['year'] == 1993].sort_values('n', ascending=False).reset_index(drop=True)
name_1993[name_1993['name'] == 'Craig']
```

```python
# The year before?
name_1991 = df[df['year'] == 1991].sort_values('n', ascending=False).reset_index(drop=True)
name_1991[name_1991['name'] == 'Craig']
```

```python
# What about 5 years before?
name_1987 = df[df['year'] == 1987].sort_values('n', ascending=False).reset_index(drop=True)
name_1987[name_1987['name'] == 'Craig']

```

```python
# What about 5 years after?
name_1997 = df[df['year'] == 1997].sort_values('n', ascending=False).reset_index(drop=True)
name_1997[name_1997['name'] == 'Craig']
```
